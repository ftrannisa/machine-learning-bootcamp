### Sorting and Renaming

```python
scores.sort_values(by = 'math score').head() # sort by column, by default the sorting happens in ascending order
```

```python
scores.sort_values(by = 'math score', ascending = False).head()
```

```python
scores['math score'].sort_values(ascending = False)[:10]
```

```python
scores.rename(columns = {'race/etnicity': 'race',
                        'parental level of education' : 'parental_education_level',
                         'test preparation course' : 'test_prep_course',
                         'math score' : 'math_score',
                         'reading score' : 'reading_score',
                         'writing score' : 'writing_score'
                        }, inplace = True) 

# inplace will make the changes in the original DataFrame
```

```python
scores.columns
```

```python
scores.head()
```

```python
scores.rename(index = {0 : 'zero', 1 : 'first'}).head() # rename the indexes
```
